```python
import pandas as pd
from src.model import load_data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.svm import LinearSVC
```

```python
data = load_data("hackathon.csv")
```

```python
data = data.query(
    "vulnerable.notnull()"
)  # remove columns where the target variable is not null
data.head()
```

```python
data.info()
```

```python
data = data.dropna(axis="columns", how="all")
```

```python
data.head()
```

```python
data.info()
```

```python
print("access_to insurance counts:", data["access_to_insurance"].value_counts())
print("income stability:", data["income_stability"].value_counts())
```

```python
# convert TRUE/FALSE dt to numeric
boolean_dt = [
    "loan_access",
    "access_to_bank_account",
    "access_to_insurance",
    "access_to_credit_card",
    "vulnerable",
    "loan_history",
]

data[boolean_dt] = data[boolean_dt].astype(int)

# options are stable/unstable for "income_stability" change to 1 or 0
data["income_stability"] = data["income_stability"].replace(
    {"Stable": 1, "Unstable": 0}
)
```

```python
# encode categorical data
categorical_data = ["region", "employment_status"]

enc = OrdinalEncoder()  # noqa: F821
enc.fit(data[categorical_data])
data[categorical_data] = enc.transform(data[categorical_data])
```

```python
print("income stability:", data["income_stability"].value_counts())
print("region:", data["region"].value_counts())
print("employment_status:", data["employment_status"].value_counts())
```

```python
data.columns
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# the independent variables set
X = data[
    [
        "income_level",
        "debt_level",
        "region",
        "loan_access",
        "access_to_credit_card",
        "access_to_insurance",
        "financial_literacy_score",
        "savings_amount",
        "income_stability",
        "loan_history",
        "late_payments_yearly",
        "bankruptcies_filed",
        #'credit_utilisation_ratio',
        # 'credit_score',
        # 'age',
        # 'access_to_bank_account',
        "employment_status",
    ]
]

# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

# calculating VIF for each feature
vif_data["VIF"] = [
    variance_inflation_factor(X.values, i) for i in range(len(X.columns))
]

print(vif_data)

# lets remove credit score as it has a VIF of 11, and age has a VIF of 6, access to bank account has a VIF of 5.5
```

```python
# drop columns with hight multicolinearity
data = data.drop(
    columns=[
        "credit_utilisation_ratio",
        "credit_score",
        "age",
        "access_to_bank_account",
    ]
)
```

```python
trainX = data.drop(columns=["vulnerable", "id"])
```

```python
data["vulnerable"].value_counts()
# need to normalise the data to have equal numbers of TRUE and FALSE
```

```python
X = trainX
y = data.vulnerable

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.5, random_state=42
)

clf = LinearSVC()
model = clf.fit(
    X_train,
    y_train,
)
score = model.score(
    X_train,
    y_train,
)
print("Score: ", score)
```
