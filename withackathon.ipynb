```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

data = pd.read_csv('./Data/hackathon.csv')

data.head()
```

```python
def load_data(filename: str):
    """
    :param filename: name of the file to load, must be in the 'Data' folder
    :return:
    """
    financial_vul = pd.read_csv('./Data/'+filename)
    return financial_vul

data = load_data('hackathon.csv')
```

```python
trainX = data.drop(columns=['vulnerable'])
```

```python
X = trainX
y= data.vulnerable

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.5,random_state =123)

clf = make_pipeline(StandardScaler(),LinearSVC(dual="auto", random_state=0, tol=1e-5))
models,predictions = clf.fit(X_train,y_train,)

print(models)
```
